```python
#Probabilidade Apriori
contRest, contWalk, contRun = 0,0,0
classes = np.array(classe)
for x in range(0, len(classes)):
    if classes[x] == 'rest':
        contRest += 1
    if classes[x] == 'walking':
        contWalk += 1
    if classes[x] == 'running':
        contRun += 1
print('Total de rest: ', contRest)
print('Total de walk: ', contWalk)
print('Total de run: ', contRun)
probRest = contRest/len(classes)
probWalk = contWalk/len(classes)
probRun = contRun/len(classes)

print('Prob rest: ',probRest*100)
print('Prob walk: ', probWalk*100)
print('Prob run', probRun*100)
```

```python
previsores = data.iloc[:,2:6].values

def load(previsor,classe,posicao):
    cont = 0
    for i in range(0, len(previsores)):
        linha = previsores[i]
        if linha[posicao] == previsor and linha[3] == classe:
            cont+=1
    return cont

```

```python
print("Frequência Diet - Kind")
print("low fat - rest: ", load("low fat", "rest", 0))
print("no fat - rest: ", load("no fat", "rest", 0))
print("low fat - walking: ", load("low fat", "walking", 0))
print("no fat - walking: ", load("no fat", "walking", 0))
print("low fat - running: ", load("low fat", "running", 0))
print("no fat - running: ", load("no fat", "running", 0))

```

```python
print("Probabilidade Condicional Diet - Kind")
print("low fat - rest: ", load("low fat", "rest", 0)/contRest*100,"%")
print("no fat - rest: ", load("no fat", "rest", 0)/contRest*100,"%")
print("low fat - walking: ", load("low fat", "walking", 0)/contWalk*100,"%")
print("no fat - walking: ", load("no fat", "walking", 0)/contWalk*100,"%")
print("low fat - running: ", load("low fat", "running", 0)/contRun*100,"%")
print("no fat - running: ", load("no fat", "running", 0)/contRun*100,"%")
```

```python
def calculaVariancia(classe):
    cont = 0
    soma = 0
    media = meanPulseByClass(classe)
    for i in range(0, len(previsores)):
        linha = previsores[i]
        if linha[3] == classe:
            soma += (linha[1] - media)**2
            cont+=1
    return soma/(cont-1)

    
```

```python
print('Variancia by rest', calculaVariancia('rest'))
print('Variancia by running', calculaVariancia('running'))
print('Variancia by walking', calculaVariancia('walking'))
```

```python
def calculaDesvio(classe):
    return mt.sqrt(calculaVariancia(classe))
```

```python
print('Desvio by rest', calculaDesvio('rest'))
print('Desvio by running', calculaDesvio('running'))
print('Desvio by walking', calculaDesvio('walking'))
```

```python
def gaussian(x, mean, stdev):
    exponent = mt.exp(-(mt.pow(x-mean,2)/(2*mt.pow(stdev,2))))
    return (1/(mt.sqrt(2*mt.pi)*stdev))*exponent
```

```python
def naive(diet, pulse, time):
    gaussRest = gaussian(pulse, meanPulseByClass('rest'), calculaDesvio('rest'))
    gaussRunning = gaussian(pulse, meanPulseByClass('running'), calculaDesvio('running'))
    gaussWalking = gaussian(pulse, meanPulseByClass('walking'), calculaDesvio('walking'))
    probDietByRest = (load(diet, "rest", 0)/contRest)
    probDietByRunning = (load(diet, "runing", 0)/contRun)
    probDietByWalking = (load(diet, "walking", 0)/contWalk)
    probTimeByRest = (load(time, "rest", 2)/contRest)
    probTimeByRunning = (load(time, "running", 2)/contRun)
    probTimeByWalk = (load(time, "walking", 2)/contWalk)
    prodRest = probRest * probDietByRest * probTimeByRest * gaussRest
    prodRunning = probRun * probDietByRunning * probTimeByRunning * gaussRunning
    prodWalking = probWalk * probDietByWalking * probTimeByWalk * gaussWalking
    if prodRest > prodRunning:
        if prodRest > prodWalking:
            return "rest"
        else:
            return "walking"
    else: 
        if prodRunning > prodWalking:
            return "running"
        else:
            return "walking"
```
